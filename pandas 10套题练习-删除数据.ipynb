```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("./exercise_data/911.csv")
df.head()
```

```python
df.info()
```

```python
title_list = df['title'].str.split(':')
res_list = list(set([item[0] for item in title_list]))

zero_df = pd.DataFrame(np.zeros((df.shape[0],len(res_list))),columns=res_list)
print(zero_df)
for core in res_list:
    zero_df[core][df['title'].str.contains(core)] = 1 # 学习一个小方法
zero_df.sum()
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'

df['cate'] = [item[0] for item in title_list]
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df.set_index('timeStamp',inplace=True)
count_by_mouth = df.resample('M').count().title
_x = count_by_mouth.index
_y = count_by_mouth.values
_x = [i.strftime("%Y%m%d") for i in _x]
plt.figure(figsize=(20,8),dpi = 80)
plt.bar(_x,_y)
plt.xticks(rotation = 30,fontsize = 13)
for x,y in enumerate(_y):
    plt.text(x,y+100, y ,ha = 'center',fontsize=13) # ha水平   vl垂直
plt.show()
```

```python
df.head()
```

```python

plt.figure(figsize=(20,8),dpi = 80)
plt.xticks(rotation = 30,fontsize = 13)
for group_name,group_data in df.groupby('cate'):
    count_by_mouth = group_data.resample('M').count().title
    _x = count_by_mouth.index
    _y = count_by_mouth.values
    _x = [i.strftime("%Y%m%d") for i in _x]
    plt.plot(_x,_y,marker = 'o',label = group_name)
    for x,y in enumerate(_y):
        plt.text(x,y+100,y,ha = "center",fontsize = 15)
plt.legend(loc = "best")
plt.show()
    
```
