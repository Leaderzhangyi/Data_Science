```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('./exercise_data/linner_data.csv',header=None)
df = data.copy()
```

```python
df.info()
```

```python
_x = df[0]
_y = df[1]
plt.scatter(_x,_y)
```

```python
# 初始化迭代次数 斜率 截距 学习率

# 根据最小二乘法计算回归代价函数
def cost_function(k,b,x_data,y_data):
    cost_error = 0
    for i in range(len(x_data)):
        cost_error += ((k*x_data[i] + b) - y_data[i]) ** 2
    return cost_error/(2.0*len(x_data))

# 利用梯度下降进行误差优化
def Gradient_Descent(x_data,y_data,era,k,b,lr):
    m = float(len(x_data))
    for i in range(era):
        b_grad =0
        k_grad = 0
        for j in range(len(x_data)):
            b_grad += -(1/m)*(y_data[j] - (k*x_data[j] + b))
            k_grad += -(1/m)*(y_data[j] - (k*x_data[j] + b))*x_data[j]
        b = b - (lr*b_grad)
        k = k - (lr*k_grad)
        # 每迭代5次画一次图
        if i % 5 == 0:
            plt.rcParams['font.sans-serif'] = 'SimHei'
            plt.title("第"+str(i // 5 + 1)+"次作图")
            plt.plot(x_data,y_data,'b.')
            plt.plot(x_data,k*x_data + b,'r-')
            plt.show()
            print("误差为：",cost_function(k,b,x_data,y_data))
       
```

```python
if __name__ == "__main__":
    era = 50
    k = 0
    b = 0
    lr = 0.0001
    first_error = cost_function(k,b,_x,_y)
    print("第一次误差为：",first_error)
    plt.plot(_x,_y,'b.')
    plt.plot(_x,k*_x + b,'r')
    plt.show()
    Gradient_Descent(_x,_y,era,k,b,lr)
```
